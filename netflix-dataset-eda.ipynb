DATASET OVERVIEW:

The file "netflix_dataset.csv" contains 7,787 records of Netflix shows and movies with 12 columns. 
It includes details such as:

show_id: Unique identifier for each show/movie

type: Whether it's a "Movie" or "TV Show"

title: Name of the show or movie

director: Name of the director (some missing values)

cast: List of main actors (some missing values)

country: Country of origin (some missing values)

date_added: Date when it was added to Netflix

release_year: Year of release

rating: Content rating (e.g., PG-13, TV-MA)

duration: Duration of the movie or number of seasons for TV shows

listed_in: Categories/genres it belongs to

description: A brief summary of the show or movie

```python
# Make counts of values in type column
type_counts = df["type"].value_counts()

type_counts
```

```python

# pie chart for Distribution of Content Types
plt.figure(figsize=(10, 6))
plt.pie(type_counts, labels=type_counts.index, autopct="%1.1f%%", colors=["#ff9999", "#66b3ff"], startangle=90,textprops={'fontsize': 16, 'color': 'black'})
plt.title("Distribution of Content Types")
plt.show()
```

```python
# Top 10 directors counts 
directors=df["director"].value_counts().head(10)
directors
```

```python
# Group by country and type
df2 = df.groupby(["country","type"]).size().reset_index(name="count")
df2
```

```python
# Get top 10 countries with most Movies and TV Shows
type_movie = df2[df2["type"] == "Movie"].nlargest(10, "count")
type_movie
```

```python
type_tvshow = df2[df2["type"] == "TV Show"].nlargest(10, "count")
type_tvshow
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart for Movies
axes[0].bar(type_movie["country"], type_movie["count"], color="skyblue")
axes[0].set_title("Top 10 Countries for Most Movies on Netflix")
axes[0].set_xlabel("Country")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis="y")


# Bar chart for TV Shows
axes[1].bar(type_tvshow["country"], type_tvshow["count"], color="orange")
axes[1].set_title("Top 10 Countries for Most TV Shows on Netflix")
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis="y")


# Show the plots
plt.tight_layout()
plt.show()
```

```python
#  Top 10 countries which have most content on Netflix
top_10 = df['country'].value_counts().nlargest(10)
top_10_df = df[df['country'].isin(top_10.index)]

# Bar Chart for Top 10 Countries
plt.figure(figsize=(12, 6))
sns.countplot(data=top_10_df, x='country', hue='type', palette='hls')
plt.title("movies and tv shows")
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```

```python
# Work on Genres 
# Split the 'listed_in' column into separate genre columns
df_genres = df["listed_in"].str.split(", ", expand=True)
df_genres
```

```python
# Rename columns to genre_1, genre_2, genre_3, etc.
df_genres.rename(columns={0:"genre_1",1:"genre_2",2:"genre_3"},inplace=True)
df_genres
```

```python
# Merge back to the original DataFrame
df_genre = pd.concat([df, df_genres], axis=1)
df_genre
```

```python
# Example Considering genre_1 as main genre
# bar graph most top 10 genres
plt.figure(figsize=(12, 6))
d=df_genre["genre_1"].value_counts().nlargest(10)
d.plot(kind="bar",x=d.index,color="plum")
plt.title("Top 10 Genre Count")
plt.xticks(rotation=45)
plt.xlabel('genre')
plt.ylabel('Count')
plt.show()
```

```python
# Rating 
certification=df.rating.value_counts()
certification
```

```python
#The "rating" column contains some incorrect values that appear to be durations
#(e.g., '74 min', '84 min', '66 min'). These should be removed or corrected.

# Define valid ratings
valid_ratings = {'PG-13', 'TV-MA', 'PG', 'TV-14', 'TV-PG', 'TV-Y', 'TV-Y7', 'R',
                 'TV-G', 'G', 'NC-17', 'NR', 'TV-Y7-FV', 'UR'}

# Replace invalid ratings with NaN
df = df[df["rating"].isin(valid_ratings)]

# Display unique values after cleaning
print(df["rating"].unique())
```

```python
# Rating 
certification=df.rating.value_counts()
certification
```

```python
# bar chart fot age certification
plt.figure(figsize=(10, 6))
sns.barplot(x=certification.index, y=certification.values, palette='husl')
plt.title("Age Certification")
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Date and Time Analysis
df.loc[:"date_added"]=pd.to_datetime(df["date_added"],format="mixed").dt.date
df
#Format Code	Meaning	Example
#%Y-%m-%d	Year-Month-Day	2023-03-28
#%d-%b-%y	Day-Month-Year (Short Month Name)	28-Mar-23
#%B %d, %Y	Full Month Name, Day, Year	March 28, 2023
#%m/%d/%Y	Month/Day/Year	03/28/2023
```

```python
df["date_added"].dtype
```

```python
# How Many Movies Added In Netflix Each Year
# Lets Extract Year From date_added column
df.loc[:,"year_added"]=pd.to_datetime(df["date_added"],errors="coerce").dt.year
```

```python
# Count the number of Movies & TV Shows per year
year_type_counts = df.groupby(["year_added", "type"]).size().unstack().fillna(0)
year_type_counts
```

```python
# Plot the data
# the year_added column is datatype of float instead of datetime
plt.figure(figsize=(12, 6))
year_type_counts.plot(kind="bar", colormap="coolwarm", figsize=(12, 6))
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.title("Number of Movies & TV Shows Added Per Year")
plt.legend(title="Type")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
# Content added per year according to countries
top_5_countries = df["country"].value_counts().nlargest(3).reset_index() 
top_5_countries
```

```python
# Filter dataset to include only top 5 countries
df_top_5 = df[df["country"].isin(top_5_countries.country)]
df_top_5
```

```python
df = df.dropna(subset=["year_added"])
df
```

```python
df["year_added"].isna().sum()
```

```python
# Group by Year and Country, then count occurrences
year_country_counts_5 = df_top_5.groupby(["year_added", "country"]).size().unstack().fillna(0)
year_country_counts_5
```

```python
# Plot the data
plt.figure(figsize=(12, 6))
year_country_counts_5.plot(kind="line", marker="o", colormap="tab10", figsize=(12, 6))
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.title("Top 3 Countries by Content Added Per Year")
plt.legend(title="Country", loc="upper left")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
# movies and tv shows added over years
df.groupby(["year_added", "type"]).size().unstack().plot(
    kind="line", marker="o", figsize=(12, 6), colormap="coolwarm"
)
plt.xlabel("Year Added")
plt.ylabel("Count")
plt.title("Trend of Movies vs TV Shows Over Time")
plt.legend(title="Type")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
# plot for content over by months
df.loc[:,"month_added"] =pd.to_datetime( df["date_added"]).dt.month
df.groupby("month_added").size().plot(
    kind="line", marker="o", figsize=(12, 6), color="red"
)
plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Content Added Per Month")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

Conclusion :

The analysis of Netflix’s content catalog reveals key insights into the platform’s content distribution and strategy. The dataset provides valuable information on content trends, such as the increasing number of shows and movies added in recent years, the diversity in content genres, and variations in ratings and duration.

Key observations include:

Regional Content Trends: The dataset highlights how Netflix diversifies its offerings based on regional preferences, with content spanning multiple countries.

Growth of Original Content: By examining the year-wise distribution, we can see how Netflix has expanded its library, particularly in original productions.

Audience Targeting: Content ratings indicate how Netflix caters to different age groups, ensuring a mix of family-friendly and mature content.

Influence of Directors and Cast: Popular directors and actors play a significant role in content selection and viewership trends.
